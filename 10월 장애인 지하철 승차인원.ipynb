```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
raw_data = pd.read_csv('../data_files/subway.csv')
raw_data
```

```python
raw_data = raw_data[raw_data['구분'] == '하차']
raw_data
```

```python
raw_data1 = pd.DataFrame({'호선':raw_data['호선'], '역명':raw_data['역명'], '하차': raw_data['10월']})
raw_data1
```

```python
raw_data1.info()
```

```python
raw_data1 = raw_data1.dropna()
```

```python
raw_data1.info()
```

```python
raw_data1 = raw_data1.reset_index()
```

```python
collections = []
for x in tqdm(range(len(raw_data1))):
    sum = 0
    for y in range(len(raw_data1)):
        if raw_data1['역명'][x] == raw_data1['역명'][y]:
            sum = raw_data1['하차'][y] + sum
    collections.append([raw_data1['역명'][x], sum])
```

```python
collections
```

```python
collections_frame = pd.DataFrame(collections, columns=['역','하차'])
collections_frame
```

```python
collections_frame = collections_frame.drop_duplicates()
collections_frame
```

```python
collections_frame = collections_frame.reset_index()
```

```python
for x in tqdm(range(len(collections_frame))):
    if collections_frame['역'][x][-1] != '역':
        collections_frame['역'][x] = collections_frame['역'][x] + '역'
collections_frame         
```

```python
import googlemaps
```

```python
my_key = "my google api key" # 구글 api 키 입력
maps = googlemaps.Client(key=my_key)
lat = []  #위도
lng = []  #경도
```

```python
i=0
for place in tqdm(collections_frame['역']):   
    i = i + 1
    try:
        geo_location = maps.geocode(place)[0].get('geometry')
        print(geo_location)
        lat.append(geo_location['location']['lat'])
        lng.append(geo_location['location']['lng'])
        

    except:
        print(i,'번 위치를 찾는데 실패')
        continue
        
```

```python
Total = collections_frame['하차'].tolist()
Total
```

```python
# 데이터프레임만들어 출력하기
map = pd.DataFrame({'역명':collections_frame ['역'], '위도':lat, '경도':lng, '하차': Total})
map
```

```python
map = map.sort_values(by='하차', ascending=False)
map
```

```python
map.info()
```

```python
map = map.reset_index()
map = map.drop(['index'], axis=1)
map
```
