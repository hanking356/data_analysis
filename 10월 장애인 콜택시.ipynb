```python
type(total_concat['승차건수'][0])
```

```python
collections = []
for x in tqdm(range(len(total_concat))):
    sum = 0
    for y in range(len(total_concat)):
        if total_concat['동/읍/면'][x] == total_concat['동/읍/면'][y]:
            sum = total_concat['승차건수'][y] + sum
    collections.append([total_concat['동/읍/면'][x], sum])
```

```python
collections
```

```python
collections_frame = pd.DataFrame(collections, columns=['목적지','이용건수'])
collections_frame
```

```python
collections_frame = collections_frame.drop_duplicates()
collections_frame
```

```python
collections_frame = collections_frame.sort_values(by='이용건수', ascending=False)
collections_frame
```

```python
collections_frame = collections_frame.reset_index()
collections_frame = collections_frame.drop(['index'], axis=1)
collections_frame
```

```python
import googlemaps
```

```python
my_key = "my google api key" # 구글 api 키 입력
maps = googlemaps.Client(key=my_key)  # my key값 입력
lat = []  #위도
lng = []  #경도
```

```python
i=0
for place in tqdm(collections_frame['목적지']):   
    i = i + 1
    try:
        geo_location = maps.geocode(place)[0].get('geometry')
        print(geo_location)
        lat.append(geo_location['location']['lat'])
        lng.append(geo_location['location']['lng'])
        

    except:
        print(i,'번 위치를 찾는데 실패')
        continue
        
```

```python
collections_frame['위도'] = lat
collections_frame['경도'] = lng
```
