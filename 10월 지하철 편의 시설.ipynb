```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
lift = pd.read_csv('../data_files/wheelchair_iift.csv')
lift
```

```python
collections = []
for x in tqdm(range(len(lift))):
    sum = 0
    for y in range(len(lift)):
        if lift['역명'][x] == lift['역명'][y]:
            sum = lift['대수'][y] + sum
    collections.append([lift['역명'][x], sum])
```

```python
collections_frame = pd.DataFrame(collections, columns=['역','대수'])
collections_frame
```

```python
collections_frame = collections_frame.drop_duplicates()
collections_frame
```

```python
collections_frame = collections_frame.reset_index()
collections_frame = collections_frame.drop(['index'], axis=1)
collections_frame
```

```python
for x in tqdm(range(len(collections_frame))):
    collections_frame['역'][x] = collections_frame['역'][x] + '역'         
```

```python
collections_frame
```

```python
collections_frame.to_excel('lift.xlsx')
```
