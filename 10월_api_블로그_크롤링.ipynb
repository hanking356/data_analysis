```python
import urllib.request
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from tqdm import tqdm
```

```python
def create_url(api_node, search_text, start_num, disp_num, sort):
    base = "https://openapi.naver.com/v1/search"
    node = "/" + api_node + ".json"
    param_query = "?query=" + urllib.parse.quote(search_text)
    param_start = "&start=" + str(start_num)
    param_disp = "&display=" + str(disp_num)
    param_sort = "&sort=" + str(sort)
    return base + node + param_query + param_start + param_disp + param_sort
```

```python
# 숫자와 한글만을 가져오는 정규식 
import re
def text_cleaning(text):
    result = re.sub('[^ 0-9ㄱ-힗]', '', text) 
    return result
```

```python
for i in range(len(result_df)):
    if type(result_df['content'][i]) != str:
        result_df['content'][i] = str(result_df['content'][i])
```

```python
for i in range(len(result_df)):
    if type(result_df['title'][i]) != str:
        result_df['title'][i] = str(result_df['title'][i])
```

```python
# 각 피처마다 데이터 전처리를 적용합니다.
result_df['title'] = result_df['title'].apply(lambda x: text_cleaning(x))
result_df['content'] = result_df['content'].apply(lambda x: text_cleaning(x))
result_df.head(5)
```

```python
# 각 피처마다 말뭉치를 생성
title_corpus = "".join(result_df['title'].tolist())
content_corpus = "".join(result_df['content'].tolist())
print(content_corpus)
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
# konlpy의 형태소 분석기로 명사 단위의 키워드를 추출
t = Okt()
nouns = t.nouns(content_corpus) #nouns는 명사를 반환, morphs는 형태소를 반환
count = Counter(nouns)
count
```

```python
# 한글자 키워드를 제거
remove_char_counter = Counter({x : count[x] for x in count if len(x) > 1})
print(remove_char_counter)
```

```python
korean_stopwords_path = "../data/korean_stopwords.txt"

# 텍스트 파일을 오픈
with open(korean_stopwords_path, encoding='utf8') as f:
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]
print(stopwords[:10])
```

```python
remove_char_counter = Counter({x : remove_char_counter[x] for x in count if x not in stopwords})
```

```python
import pytagcloud
```

```python
ranked_tags = remove_char_counter.most_common(40)
taglist = pytagcloud.make_tags(ranked_tags, maxsize=80)
pytagcloud.create_tag_image(taglist, 'wordcloud.jpg', size=(900, 600), fontname='NanumGothic', rectangular=False)
```
