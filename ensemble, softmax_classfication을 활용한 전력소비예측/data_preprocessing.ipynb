```python
raw_data['time'] = 0
```

```python
for x in tqdm(range(85)):
    for y in range(x*24,8+(x*24)):
        raw_data['time'][y] = 0
    for y in range(8+(x*24),24+(x*24)):        
        raw_data['time'][y] = 1 
```

```python
raw_data.pivot_table(index='time', aggfunc=[np.mean, np.std])
```

```python
raw_data2 = pd.get_dummies(raw_data,columns=['time'])
raw_data2
```
