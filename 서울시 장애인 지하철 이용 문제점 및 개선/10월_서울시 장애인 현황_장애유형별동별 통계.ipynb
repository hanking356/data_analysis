```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
raw_data = pd.read_csv('../data_files/people.txt', delimiter = '\t')
raw_data
```

```python
raw_data.drop(columns='기간', inplace=True)
raw_data.drop(columns='합계.1', inplace=True)
raw_data.drop(columns='합계.2', inplace=True)
raw_data.drop(columns='지체.1', inplace=True)
raw_data.drop(columns='지체.2', inplace=True)
raw_data.drop(columns='뇌병변.1', inplace=True)
raw_data.drop(columns='뇌병변.2', inplace=True)
raw_data.drop(columns='시각.1', inplace=True)
raw_data.drop(columns='시각.2', inplace=True)
raw_data.drop(columns='청각.1', inplace=True)
raw_data.drop(columns='청각.2', inplace=True)
raw_data.drop(columns='언어.1', inplace=True)
raw_data.drop(columns='언어.2', inplace=True)
raw_data.drop(columns='지적장애.1', inplace=True)
raw_data.drop(columns='지적장애.2', inplace=True)
raw_data.drop(columns='자폐성.1', inplace=True)
raw_data.drop(columns='자폐성.2', inplace=True)
raw_data.drop(columns='정신장애.1', inplace=True)
raw_data.drop(columns='정신장애.2', inplace=True)
raw_data.drop(columns='신장장애.1', inplace=True)
raw_data.drop(columns='신장장애.2', inplace=True)
raw_data.drop(columns='심장장애.1', inplace=True)
raw_data.drop(columns='심장장애.2', inplace=True)
raw_data.drop(columns='호흡기.1', inplace=True)
raw_data.drop(columns='호흡기.2', inplace=True)
raw_data.drop(columns='간.1', inplace=True)
raw_data.drop(columns='간.2', inplace=True)
raw_data.drop(columns='안면.1', inplace=True)
raw_data.drop(columns='안면.2', inplace=True)
raw_data.drop(columns='장루요루.1', inplace=True)
raw_data.drop(columns='장루요루.2', inplace=True)
raw_data.drop(columns='뇌전증.1', inplace=True)
raw_data.drop(columns='뇌전증.2', inplace=True)
```

```python
raw_data = raw_data.drop([0,1])
raw_data
```

```python
raw_data = raw_data.reset_index()
raw_data
```

```python
for x in tqdm(range(len(raw_data))):
    raw_data['합계'][x] = raw_data['합계'][x].replace(",", "")
    raw_data['청각'][x] = raw_data['청각'][x].replace(",", "")
    raw_data['언어'][x] = raw_data['언어'][x].replace(",", "")
    raw_data['안면'][x] = raw_data['안면'][x].replace(",", "")
    raw_data['합계'][x] = int(raw_data['합계'][x])
    raw_data['청각'][x] = int(raw_data['청각'][x])
    raw_data['언어'][x] = int(raw_data['언어'][x])
    raw_data['안면'][x] = int(raw_data['안면'][x])
raw_data    
```

```python
raw_data['총 인원'] = (raw_data['합계'] - raw_data['청각'] - raw_data['언어'] - raw_data['안면'])*31
```

```python
raw_data
```

```python
raw_data.drop(columns='합계', inplace=True)
raw_data.drop(columns='청각', inplace=True)
raw_data.drop(columns='언어', inplace=True)
raw_data.drop(columns='안면', inplace=True)
raw_data = raw_data.drop(['index'], axis=1)
raw_data
```

```python
raw_data = raw_data.sort_values(by='총 인원', ascending=False)
raw_data
```

```python
raw_data = raw_data.reset_index()
raw_data
```
