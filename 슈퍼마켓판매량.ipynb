```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
super = pd.read_csv('../data/supermarket_sales - Sheet1.csv', index_col=0, encoding='utf-8')
super
```

### Product line에서 Health and beauty 항목만 가져오기

```python
super2 = super[super['Product line'] == 'Health and beauty']
super2
```

### Gender column에서 Male과 Female의 항목 갯수 구하기

```python
len(super2[super2['Gender'] == 'Male'])
```

```python
len(super2[super2['Gender'] == 'Female'])
```

### Gender를 index로 한 pivot_table을 생성해서 super3에 저장

```python
super3 = super2.pivot_table(index='Gender', aggfunc=np.sum)
super3
```

### 파생변수 gnum을 만들어 남자와 여자의 항목 갯수를 넣기

```python
super3['gnum'] = [64, 88]
super3
```

### 남자와 여자 사람수 대비 healthy and beauty의 소비 금액에 대한 pie chart

```python
import matplotlib.pyplot as plt

f, ax  = plt.subplots(1, 2, figsize=(16, 8)) #1행 2열

super3['gnum'].plot.pie(explode=[0, 0.05], ax = ax[0], autopct='%1.1f%%')
ax[0].set_title('gender')

super3['Total'].plot.pie(explode=[0, 0.05], ax = ax[1], autopct='%1.1f%%')
ax[1].set_title('total cost')
```
